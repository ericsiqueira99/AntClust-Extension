# change hyperparameters

```python
# similarity function
min = 0
max = 100
f_sim = [similarity_1d(min, max)]

# define rules
rules = labroche_rules()

# parameter search
results = []
alpha_range = np.arange(100,200, 1)
betta_range = np.arange(0.01, 1, 0.05)
for alpha in alpha_range:
    print()
    print('alpha is: ', alpha)
    print()
    betta_tmp_list = []
    for betta in betta_range:
        # define Antclust
        ant_clust = AntClust(
            data,
            f_sim,
            rules,
            store_computed_similaritys=True,
            alpha_ant_meeting_iterations=alpha,
            betta_template_init_meetings=betta,
            nest_shrink_prop=0.2,
            nest_removal_prop=0.3,
        )

        # find clusters
        ant_clust.find_clusters()

        # get clutsers
        betta_tmp_list.append(ant_clust.get_clusters())
    results.append(betta_tmp_list)
```

### generating heatmap

```python
# calc number of clusters found

num_clusters = []
for alpha in results:
    tmp = []
    for betta in alpha:
        tmp.append(np.max(np.array(betta)))
    num_clusters.append(tmp)
```

```python
# 1. Import Modules
import numpy as np
import seaborn as sns
import matplotlib.pylab as plt
plt.style.use("seaborn")
 
# 2. print data
print("Our dataset is : ")
print(num_clusters)
 
# 3. Plot the heatmap
plt.figure(figsize=(10,10))
heat_map = sns.heatmap( num_clusters, linewidth = 1 , annot = True)
plt.title( "NumClusters" )
plt.show()
```
