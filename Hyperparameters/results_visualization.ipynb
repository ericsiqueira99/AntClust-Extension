# Visualization of results

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the CSV file into a DataFrame
df = pd.read_csv('results.csv')
df
```

```python
# Assuming your CSV file has columns named 'column1', 'column2', and 'column3'
# Replace them with the actual column names from your CSV file
fixed_ari_name = "Fixed ari"
GA_ari_name = "GA ari"
PSO_ari_name = "PSO ari"
fixed_ari_val= df[fixed_ari_name]
GA_ari_val = df[GA_ari_name]
PSO_ari_val = df[PSO_ari_name]

# Assuming the range is 0 to 10
x = df["Num clusters"]

# Plotting
plt.plot(x, fixed_ari_val, label="Fixed params")
plt.plot(x, GA_ari_val, label= "GA params")
plt.plot(x, PSO_ari_val, label="PSO params")

# Adding labels and legend
plt.xlabel('Number of clusters')
plt.ylabel('ARI score')
plt.title('Parameters comparisson')
plt.legend()

# Display the plot
plt.show()
```
