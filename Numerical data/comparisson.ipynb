```python
import matplotlib.pyplot as plt

# Sample data
antclust = [(4, 0.544441), (13, 0.399096), (30, 0.503367), (64, 0.519198)]
#best_kmean= [(4, 0.730238), (13, 0.371114), (30, 0.501563), (64, 0.731170)]
class_kmean= [(4, 0.730238), (13, 0.371114), (30, 0.491425), (64, 0.664962)]

# Unzip the lists of tuples
keys1, values1 = zip(*antclust)
keys2, values2 = zip(*class_kmean)

# Plotting
plt.plot(keys1, values1, 'o--',label='AntClust')
plt.plot(keys2, values2, 'o--', label='Best_kmeans')

# Adding labels and title
plt.xlabel('Number of features')
plt.ylabel('ARI')
plt.title('AntClust vs k-means (k=classes)')

# Adding legend
plt.legend()

# Display the plot
plt.show()
```
